```python
# Now each dataFrame are in sequence
X_Y_train_l = pd.read_csv('titanic_x_y_train.csv')
X_test_l = pd.read_csv('titanic_x_test.csv')

X_Y_train_l.head()
```

# DataPreprocessing over titanic dataset

```python
# Check is there is any total null values in the columnFeature
X_Y_train_l.isnull().sum() #return the total values from each col
```

```python
# Deleting those features which are not useful for training the model
# Are : Name,Ticket,Fare,Cabin
del X_Y_train_l["Name"]
del X_Y_train_l["Ticket"]
del X_Y_train_l["Cabin"]
del X_Y_train_l["Fare"]

X_Y_train_l.head()
```

```python
# Doing same in testing data
del X_test_l["Name"]
del X_test_l["Ticket"]
del X_test_l["Cabin"]
del X_test_l["Fare"]
X_test_l.head()
```

```python
# Will replace the Nan values from the mean of other datapoints in that features (Age)
# If nan value is there and other features are alphabets then would drop that row as median is not possible there (Embark)
print(X_Y_train_l.isnull().sum())
print(X_Y_train_l.shape)

X_Y_train_l = X_Y_train_l.fillna(X_Y_train_l.median())
print(X_Y_train_l.isnull().sum())
print(X_Y_train_l.shape)
#Same above for test
X_test_l = X_test_l.fillna(X_test_l.median())
print(X_test_l.isnull().sum())

X_Y_train_l.dropna(inplace=True) #inplade true means directly implimented over the dataFrame passed , no knew dataframe is returned
print(X_Y_train_l.isnull().sum())
print(X_Y_train_l.shape)
```

```python
X_train_l = X_Y_train_l.iloc[:,:X_Y_train_l.shape[1]-1]
print(X_train_l.shape)
Y_train_l = X_Y_train_l.iloc[:,-1]
print(Y_train_l.shape)
```

## Label Encoding for strings -> Sex

```python
# For training data
def getNumberedGender(str):
    if str=="male":
        return 1
    else:
        return 2
X_train_l['gender']=X_train_l["Sex"].apply(getNumberedGender) #Will impliment on Sex feature/col and return all the output from the function to newColumn called Gender in the dataFrame
print(X_train_l)
del X_train_l["Sex"] #After encoding of sex into gender col, will delete the whole Sex column
X_train_l.head()
```

```python
# Same above procedure with testing data
X_test_l['gender']=X_test_l["Sex"].apply(getNumberedGender)
print(X_test_l)
del X_test_l["Sex"]
print(X_test_l.isnull().sum())
X_test_l.head()
```

## Label Encoding for strings -> Embark

```python
def getNumberEmbarked(str):
    if str=='C':
        return 1
    elif str=='Q':
        return 2
    else:
        return 3
    
# for training data    
X_train_l['Embarked']=X_train_l["Embarked"].apply(getNumberEmbarked) #Overwrited over the old stringed Embark from encoded embark 
print(X_train_l)
X_test_l['Embarked']=X_test_l["Embarked"].apply(getNumberEmbarked)
X_test_l
```

## Now training over Deicion Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train_l, Y_train_l)
Y_pred_l = clf.predict(X_test_l)
clf.score(X_train_l, Y_train_l) , clf.score(X_test_l, Y_test)
```

## Now training over RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth = 5, random_state=5)# Random_state given bcz while running each time it will create new set of trees, so making the same trees to be built while running each time in the forest, given randomState 
clf.fit(X_train_l, Y_train_l)
clf.score(X_train_l,Y_train_l),clf.score(X_test_l,Y_test)
```
